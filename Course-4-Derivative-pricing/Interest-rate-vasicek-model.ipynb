```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def vasicek(r0, K, theta, sigma, T, N, M):
    dt = T / N
    rates = np.zeros((N, M))
    rates[0, :] = r0
    for j in range(M):
        for i in range(1, N):
            dr = (
                K * (theta - rates[i - 1, j]) * dt
                + sigma * np.sqrt(dt) * np.random.normal()
            )
            rates[i, j] = rates[i - 1, j] + dr
    return rates
```

```python
np.random.seed(42)
M = 25  # Number of paths for MC
N = 250  # Number of steps
T = 0.25  # Maturity
r0 = 2.85
K = 0.15
theta = 0.0255
sigma = 0.85
t = np.linspace(0, T, N)

rates = vasicek(r0, K, theta, sigma, T, N, M)
min(rates[-1])
```

```python
for j in range(M):
    plt.plot(t, rates[:, j])

plt.xlabel("Time $t$", fontsize=14)
plt.ylabel("$r(t)$", fontsize=14)
plt.title("Vasicek Paths", fontsize=14)
axes = plt.gca()
axes.set_xlim([0, T])
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.show()
```
