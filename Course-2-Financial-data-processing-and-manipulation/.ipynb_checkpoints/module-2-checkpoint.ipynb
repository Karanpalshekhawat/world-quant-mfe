```python
import datetime
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas_datareader.data as web
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
#yf.pdr_override()
```

```python
start = datetime.date.today() - datetime.timedelta(days=5 * 365)
end = datetime.date.today()
```

```python
df = web.DataReader(["sp500", "NASDAQCOM", "CBBTCUSD"], 'fred', start, end)
```

```python
df.head()
```

```python
df = df.dropna()
df["SP500"] = np.log(df.sp500) - np.log(df.sp500.shift(1))
df["NASDAQ"] = np.log(df.NASDAQCOM) - np.log(df.NASDAQCOM.shift(1))
df["Bitcoin"] = np.log(df.CBBTCUSD) - np.log(df.CBBTCUSD.shift(1))
df = df.iloc[1:, 3:]
```

```python
df.describe()
# we can see bitcoin is obviously more volatile
```

```python
df["2019-05-01":"2019-05-31"].plot()
# nice way to give indexes for date time format, (did not knew we can give in YYYY-MM-DD format)
```

```python
df.std()
```

```python
ax1 = df.plot(figsize=(15, 3), y="SP500", title="Figure 1: S&P 500 Daily Returns")
ax2 = df.plot(figsize=(15, 3), y="Bitcoin", title="Figure 2: Bitcoin Daily Returns")

ax1.set_ylim(-0.5, 0.4)
ax2.set_ylim(-0.5, 0.4)
# more up and down in bitcoin returns
```

Sharp ratio is excess return over risk free return divided by the volatility which is very important metric to compare risk-reward

```python
Sharpe_Ratio_SP500 = df["SP500"].mean() / df["SP500"].std()
Sharpe_Ratio_SP500
```

```python
Sharpe_Ratio_Bitcoin = df["Bitcoin"].mean() / df["Bitcoin"].std()
Sharpe_Ratio_Bitcoin
```

## Semi-variance is the concept to look into returns when it is negative, and compute standard deviation for only those, it is done to actually look into how much can we loose as most of the investors are more concerened about the downside risk than the upside risk, in short, semi-variance uses only the one side of the returns either positive or negative

```python
df[df["SP500"] < df["SP500"].mean()]["SP500"].std()
```

```python
df[df["Bitcoin"] < df["Bitcoin"].mean()]["Bitcoin"].std()
```
