### Straddle : Where you want to trade volatility by buying both call and put at the same strike and payoff will be there if the underlying price moves in any direction, that is what expecting volatility means

```python
myCall = {"op_type": "c", "strike": 100, "tr_type": "b", "op_pr": 2.49}
myPut = {"op_type": "p", "strike": 100, "tr_type": "b", "op_pr": 2.09}
op_list = [myCall, myPut]
op.multi_plotter(spot=100, spot_range=10, op_list=op_list)
```

```python
myCall = Option(
    european=True, kind="call", s0=100, k=100, t=30, sigma=0.20, r=0.05, dv=0
)
print(myCall)
```

```python
callBS = round(myCall.getPrice(), 4) #analytical
callBT = round(myCall.getPrice(method="BT", iteration=5000), 4) # Numberical 
callMC = round(myCall.getPrice(method="MC", iteration=500000), 4) # Numerical
[callBS, callBT, callMC]
```
