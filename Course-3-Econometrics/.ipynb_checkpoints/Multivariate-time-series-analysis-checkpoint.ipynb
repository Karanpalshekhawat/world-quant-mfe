```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
dataset_path = r"https://s3.amazonaws.com/assets.datacamp.com/production/course_5226/datasets/ch4_meat.csv"
df = pd.read_csv(dataset_path, parse_dates=["date"])
df = df[['date', 'beef', 'veal', 'pork', 'lamb_and_mutton']]
df = df.set_index('date')
```

```python
df.plot(figsize=(10, 6))
plt.legend(fontsize="large")
plt.xlabel("Date")
plt.ylabel("Monthly Meat Production (t)");
```
